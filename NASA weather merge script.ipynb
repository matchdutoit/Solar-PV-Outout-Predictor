```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('data/readyforweath.csv')
weath = pd.read_csv('data/weather_unique.csv')
```

```python
df.info(verbose=1)
```

```python
weath['year']=weath['year'].astype(int)
```

```python
new_weath = pd.DataFrame(columns=weath.columns)
new_weath
```

```python
for i in df.index:
    lat = df.at[i,'lat']
    lon = df.at[i,'lon']
    year = df.at[i,'year']
    dflist = [lat,lon,year]
    for j in weath.index:
#         print('looking for match')
        la = weath.at[j,'lat']
        lo = weath.at[j,'lon']
        ye = weath.at[j,'year']
        weathlist = [la,lo,ye]
        if dflist == weathlist:
            print(f"match! {i}")
            new_weath = new_weath.append(weath.loc[j,:])
            print('APPENDED!!!')
        
```
